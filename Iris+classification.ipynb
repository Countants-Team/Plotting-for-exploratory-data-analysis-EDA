Plotting for exploratory data analysis (EDA)

What is EDA?

EDA is a task of analysing your dataset using simple tools like statistics, Linear Algebra and other techniques to understand what is dataset before we model and do actual machine learning.But this is extremely important stage.For any given problem what first thing we do is EDA. We can use different plotting techniques to understand our dataset.

We can start this with the simple example of Iris Dataset:
Iris dataset : https://en.wikipedia.org/wiki/Iris_flower_data_set  (3 flowers of Iris family)

Objective : To classify the flower as belonging to the one of 3 classes given the 4 features.
            (Petal length, Petal width, Sepal length and Sepal width)
            

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

#Load Iris Dataset into a Pandas dataFrame.
iris = pd.read_csv("iris.csv")
```

```python
# How many data points and features?
print(iris.shape)
```

```python
# What are the column names in dataset?
print(iris.columns)
```

```python
# How many data points for each class are present?
iris['species'].value_counts()
```

```python
#2-D scatter plot
iris.plot(kind='scatter',x='sepal_length',y='sepal_width');
plt.show()
# We can not make much sense out of it
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(iris,hue="species",size =4)\
   .map(plt.scatter,"sepal_length","sepal_width")\
   .add_legend();
plt.show();
```

```python
# We can see that blue points can be easily separated from red and green
# But red and green data points can not be easily separated
```

OBSERVATION(S)
1. Using Sepal_width & sepal_length features, we can distinguish Satosa flowers from others.
2. Separating Versicolor and virginica is much harder as they have considerable overlap.

```python
# Pair Plots
plt.close();
sns.set_style('whitegrid');
sns.pairplot(iris,hue='species',size=3);
plt.show()
```

OBSERVATIONS:
1. Petal length and petal width are the most important features to classify various flower types.
2. While Setosa is linearly separable, Virginica and Virsicolor have some overlap
3. We can find the 'line' and 'if else' conditions to build the simple classification model.
